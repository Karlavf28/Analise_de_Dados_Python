```python
#1 - Abra o arquivo csv covid_brasil.csv em um DataFrame Pandas.
import pandas as pd
df_covid_brasil = pd.read_csv("covid_brasil.csv")
df_covid_brasil
```

```python
#2 - Determine qual cidade do Brasil apresenta o menor número de novos casos de covid-19 no Brasil.

df_casosnovos = df_covid_brasil.loc[:,["casosNovos","municipio"]]
df_casosnovos = df_casosnovos.sort_values(by="casosNovos",ascending=True)
df_casosnovos

menor = df_casosnovos.min()

print("Cidade com menor numero de casos novos:")

df_casosnovos.loc[df_casosnovos['casosNovos']==menor[0]]
```

```python
#3 - Determine qual cidade do Brasil apresenta o maior número de novos óbitos de covid-19 no Brasil.

df_obitosnovos = df_covid_brasil.loc[:,["obitosNovos","municipio"]]
df_obitosnovos = df_obitosnovos.sort_values(by="obitosNovos",ascending=False)
df_obitosnovos

maior = df_obitosnovos.max()

print("Cidade com maior numero de Obitos novos:")

df_obitosnovos.loc[df_obitosnovos['obitosNovos']==maior[0]]
```

```python
#4 - Adicione uma coluna ao DataFrame que contenha o número de casos acumulados por 100 mil habitantes
df_covid_brasil["Acumulados_100mil"] = df_covid_brasil["casosAcumulado"] / 100000
df_covid_brasil
```

```python
#5- Determine qual a cidade que apresenta o maior número de casos acumulados por 100 mil habitantes.

df_acumulados100 = df_covid_brasil.loc[:,["Acumulados_100mil","municipio"]]
df_acumulados100 = df_acumulados100.sort_values(by="Acumulados_100mil",ascending=False)
df_acumulados100

maior = df_acumulados100.max()

print("Cidade com maior numero de casos acumulados a cada 100 habitantes:")

df_acumulados100.loc[df_acumulados100['Acumulados_100mil']==maior[0]]
```

```python
#6 - Adicione uma coluna ao DataFrame que contenha o número de óbitos acumulados por 100 mil habitantes
df_covid_brasil["ObitosAcumulados_100mil"] = df_covid_brasil["obitosAcumulado"] / 100000
df_covid_brasil
```

```python
#7- Determine qual a cidade que apresenta o menor número de óbitos acumulados por 100 mil habitantes.

df_obitos100 = df_covid_brasil.loc[:,["ObitosAcumulados_100mil","municipio"]]
df_obitos100 = df_obitos100.sort_values(by="ObitosAcumulados_100mil",ascending=True)
df_obitos100

menor = df_obitos100.min()

print("Cidade com menor numero de obitos acumulados por 100mil habitantes")

df_obitos100.loc[df_obitos100['ObitosAcumulados_100mil']==menor[0]]
```

```python
# 8 - Salve o DataFrame com as duas colunas adicionadas em um arquivo excel
df_covid_brasil.to_excel('covid_brasil_v2.xlsx', index=False)
df_covid_brasil
```

```python
# 9 - Faça uma tabela contendo os dados de covid-19 apenas do estado de Minas Gerais e salve em um arquivo csv separado por ponto e vírgula.

df_MG = df_covid_brasil.loc[df_covid_brasil['estado']=='MG']
df_MG.to_csv("covid_Minas.csv", sep = ';')         
df_MG
```

```python
#10 - Determine qual o número de casos acumulados de covid-19 da região metropolitana de Belo Horizonte.

df_MG = pd.read_csv("covid_Minas.csv", sep = ';') 
df_BH = df_MG.loc[df_MG['municipio']=='Belo Horizonte']
df_MET_BH = df_MG.loc[df_MG['codRegiaoSaude']==31016.0]
df_MET_BH.sum()
print("Casos acumulados de covid-19 da região metropolitana de BH:")
df_MET_BH.sum()[11]
```

```python
#11 - Faça uma tabela com as cidades de MG que apresentaram mais de 100 novos casos de covid-19 nesse dia.
# hoje: 25/02/2020
df_MG_Hoje = df_MG.loc[df_MG['casosNovos']>100]
df_MG_Hoje
```

```python
# 12 - Construa um grafico de pontos relacionando as seguintes colunas do DataFrame com os dados de todo o Brasil
#casosAcumulado X populacaoTCU2019
#casos/habitante X populacaoTCU2019

import matplotlib.pyplot as plt
x = df_covid_brasil['casosAcumulado']
y = df_covid_brasil['populacaoTCU2019']
y1 = df_covid_brasil['casosAcumulado']/df_covid_brasil['populacaoTCU2019']

plt.figure()
plt.scatter(x,y, label='casosAcumulado X populacaoTCU2019',color='r')
plt.scatter(x,y1,label='casos/habitante X populacaoTCU2019',color='b')
plt.xlabel('Casos Acumulados')
plt.ylabel('Populacao')
plt.title('Covid-19 - Brasil')
plt.legend()
plt.show()
```
