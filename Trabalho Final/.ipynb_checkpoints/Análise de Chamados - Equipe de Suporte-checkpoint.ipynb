```python
#1 - Abrindo o arquivo de chamados do sistema Mobile no formato xlsx em um DataFrame Pandas.
import pandas as pd
df_mobile = pd.read_excel("mobile_19_03_21.xlsx")
df_mobile
```

```python
#2 - Abrindo o arquivo de chamados do sistema Desktop no formato xlsx em um DataFrame Pandas.
import pandas as pd
df_desktop = pd.read_excel("desktop_19_03_21.xlsx")
df_desktop
```

```python
#Total de chamados do mobile por tipo de erro
df_tipo_erro = df_mobile.loc[:,["Tipo de Erro"]]
freq_tipo = df_tipo_erro['Tipo de Erro'].value_counts() 

total_chamados_mobile = freq_tipo.sum()


#Total de chamados Desktop
df_tipo_erro_desktop = df_desktop.loc[:,["Tipo de Erro"]]
freq_tipo2 = df_tipo_erro_desktop['Tipo de Erro'].value_counts() 

total_chamados_desktop = freq_tipo2.sum()


#Gráfico proporcional de chamados Mobile e Desktop

slices = [total_chamados_desktop,total_chamados_mobile ]
activities = ['Desktop' , 'Mobile']
cols = ['b', 'r']

plt.pie(slices,labels=activities,colors=cols, startangle=90,shadow=True, explode=(0,1),autopct='%1.1f%%')

plt.title("Chamados abertos N2")
plt.show()

print("Total Mobile:",total_chamados_mobile)
print("Total Desktop:", total_chamados_desktop)
print("Total Geral:", total_chamados_mobile+total_chamados_desktop )
```

```python
#Mostrar o total de chamados do mobile por tipo de erro
df_tipo_erro = df_mobile.loc[:,["Tipo de Erro"]]
freq_tipo = df_tipo_erro['Tipo de Erro'].value_counts() 

import matplotlib.pyplot as plt

plt.bar(freq_tipo.index[0],freq_tipo[0],label="OUTRO",color='c', width=.5)
plt.bar(freq_tipo.index[1],freq_tipo[1],label="PROC",color='g', width=.5)
plt.bar(freq_tipo.index[2],freq_tipo[2],label="ASC ESTRUTURA",color='y', width=.5)
plt.xlabel('Tipos de Erro')
plt.ylabel('Número de chamados abertos')
plt.title("Nº Chamados Mobile ")

plt.show()
print(freq_tipo) 

total_chamados_mobile = freq_tipo.sum()
print("Total Mobile:",total_chamados_mobile)


```

```python
import numpy as np

#Mostrar o total de chamados do mobile por técnico
df_tecnico = df_mobile.loc[:,["Técnico N2"]]
df_tecnico
freq_tec = df_tecnico['Técnico N2'].value_counts() 

#print(freq_tec)

anderson_m = freq_tec[0]
valeria_m = freq_tec[1]

#Mostrar o numero de chamados por técnico no Destopk 
df_tecnico2 = df_desktop.loc[:,["Técnico N2"]]
df_tecnico2
freq_tec2 = df_tecnico2['Técnico N2'].value_counts() 

#print(freq_tec2) 

anderson_d = freq_tec2[1]
valeria_d = freq_tec2[0]

#Gráfico da distribuição dos chamados

valeria_chamados = [valeria_m,valeria_d]
anderson_chamados = [anderson_m,anderson_d]

barWidth = 0.15
plt.figure(figsize=(6,3))

r1 = np.arange(len(valeria_chamados))
r2 = [x + barWidth for x in r1]

plt.bar(r1, valeria_chamados, color = '#6A5ACD', width=barWidth, label = 'Valeria')
plt.bar(r2, anderson_chamados, color = '#00BFFF', width=barWidth, label = 'Anderson')

plt.xlabel('Chamados')
plt.xticks([r+barWidth for r in range (len (valeria_chamados))], ['Mobile','Desktop'])
plt.ylabel('Nº de chamados')
plt.title('Distribuição de chamados ao N2')

plt.legend()
plt.show()

#Numero de chamados Mobile sem resposta por técnico
df_p_resposta = df_mobile.loc[:,["Técnico N2","Primeira resposta N2"]]

df_p_resposta_v = df_p_resposta.loc[df_p_resposta['Técnico N2']=='Valeria']
df_sr_v = df_p_resposta_v.loc[df_p_resposta_v['Primeira resposta N2']=='SR']

df_p_resposta_a = df_p_resposta.loc[df_p_resposta['Técnico N2']=='Anderson']
df_sr_a = df_p_resposta_a.loc[df_p_resposta_a['Primeira resposta N2']=='SR']


total_sr_a  = df_sr_a['Primeira resposta N2'].value_counts() 
total_sr_v = df_sr_v['Primeira resposta N2'].value_counts() 


#Numero de chamados Desktop sem resposta por técnico
df_p_resposta2 = df_desktop.loc[:,["Técnico N2","Primera resposta N2"]]

df_p_resposta_v2 = df_p_resposta2.loc[df_p_resposta2['Técnico N2']=='Valeria']
df_sr_v2 = df_p_resposta_v2.loc[df_p_resposta_v2['Primera resposta N2']=='SR']

df_p_resposta_a2 = df_p_resposta2.loc[df_p_resposta2['Técnico N2']=='Anderson']
df_sr_a2 = df_p_resposta_a2.loc[df_p_resposta_a2['Primera resposta N2']=='SR']


total_sr_a2  = df_sr_a2['Primera resposta N2'].value_counts() 
total_sr_v2 = df_sr_v2['Primera resposta N2'].value_counts() 


#Mostrar o SLA do Mobile de primeiro contato de cada um dos técnicos com o usuário.
df_p_resposta_a = df_p_resposta_a.loc[df_p_resposta_a['Primeira resposta N2']!='SR']
df_p_resposta_v = df_p_resposta_v.loc[df_p_resposta_v['Primeira resposta N2']!='SR']

sla_a = df_p_resposta_a['Primeira resposta N2'].mean()
sla_v = df_p_resposta_v['Primeira resposta N2'].mean()

#Mostrar o SLA do Desktop de primeiro contato de cada um dos técnicos com o usuário.
df_p_resposta_a2 = df_p_resposta_a2.loc[df_p_resposta_a2['Primera resposta N2']!='SR']
df_p_resposta_v2 = df_p_resposta_v2.loc[df_p_resposta_v2['Primera resposta N2']!='SR']

sla_a2 = df_p_resposta_a2['Primera resposta N2'].mean()
sla_v2 = df_p_resposta_v2['Primera resposta N2'].mean()

print("Mobile: ") 
print("Anderson:",anderson_m , "chamados / ",total_sr_a[0],"Sem retorno /", sla_a," dias de SLA") 
print("Valeria:", valeria_m , "chamados / ",total_sr_v[0],"Sem retorno /", sla_v," dias de SLA")  
print("\nDesktop: ") 
print("Anderson:",anderson_d,"chamados / ",total_sr_v2[0],"Sem retorno /", sla_a2," dias de SLA")   
print("Valeria:", valeria_d, "chamados / ",total_sr_v2[0],"Sem retorno /", sla_v2," dias de SLA")   
```

```python
#Mostrar o numero de chamados por técnico no mobile caso nao houvesse os erros PROC e ESTRUTURA ASC

df_tecnico_sem_erros = df_mobile.loc[:,["Técnico N2", "Tipo de Erro"]]
df_tecnico_sem_erros = df_tecnico_sem_erros.loc[df_tecnico_sem_erros['Tipo de Erro']=='OUTRO']
df_tecnico_sem_erros

freq_sem_erros = df_tecnico_sem_erros['Técnico N2'].value_counts() 
print(freq_sem_erros) 
```

```python
#Quantos chamados foram resolvidos no dia 19/03/2021  por cada um dos técnicos no total?

df_fechados_mobile = df_mobile.loc[:,["Técnico N2", "Fechado em","Tempo N2"]]
df_fechados_mobile = df_fechados_mobile.loc[df_fechados_mobile['Fechado em']!='EA']
df_fechados_mobile

df_fechados_desktop = df_desktop.loc[:,["Técnico N2", "Fechado em","Tempo N2"]]
df_fechados_desktop = df_fechados_desktop.loc[df_fechados_desktop['Fechado em']!='EA']
df_fechados_desktop

total_fechados  = df_fechados_mobile['Técnico N2'].value_counts() + df_fechados_desktop['Técnico N2'].value_counts()
print(total_fechados) 

Media_fechados_dia = total_fechados.mean()
print("Média de chamados fechados/ atendidos por técnico por dia", Media_fechados_dia)

```

```python
#Qual a média de tempo que os chamados que foram fechados no dia 19/03 permaneceram em tratativa?

media_mobile = df_fechados_mobile['Tempo N2'].mean()
media_desktop = df_fechados_desktop['Tempo N2'].mean()

print("Desktop", media_mobile)
print("Mobile", media_desktop)
```

```python
#Partindo da hipótese que cada técnico atende uma média de 17 chamados por dia, e não houver nenhum problema 
# fora do padrão como (PROC e ESTRUTURA ASC), quantos técnicos seriam necessários para atender 
#todos os chamados  com o SLA de um dia? Considere a média de X chamados abertos por dia.


total_de_tecnicos = total_geral_chamados_sem_erro / Media_fechados_dia
print("Capacidade ideal",total_de_tecnicos )

total_de_tecnicos_max = total_geral_chamados_com_erro / Media_fechados_dia
print("Capacidade ideal",total_de_tecnicos_max )
```

```python
#Tratando os dados que serão analisados pelo HCA - Análise exploratória
import scipy
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
df_analise =  df_mobile.loc[:,["Numero","Tempo N1", "Primeira resposta N2", "Tempo N2", "Tempo N3"]]
df_analise = df_analise.loc[df_analise['Primeira resposta N2']!='SR']
df_analise
```

```python
Z = linkage(df_analise, 'ward')
fig = plt.figure(figsize=(10,10))
dn = dendrogram(Z,orientation="right", labels=[(df_analise.index[i]) for i in range(len(df_analise))])
```

```python
from bokeh.plotting import figure, output_file, show

datas = df_mobile['Aberto Em']
tempo = df_mobile['Tempo N2']

output_file("bokeh_exemplo_dispersao.html", title="Exemplo Gráfico de Dispersão")

fig2 = figure(title="exemplo grafico de dispersao", x_axis_label='datas', 
              y_axis_label='tempo')

fig2.circle(datas, tempo, size=5, alpha=0.5)

show(fig2)
```
